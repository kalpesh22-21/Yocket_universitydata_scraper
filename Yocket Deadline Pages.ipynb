```python
import csv
import time
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import ElementClickInterceptedException
from selenium.common.exceptions import NoSuchElementException
```

```python
# Provide the path of your latest browser driver.
driver_path = 'C:\Program Files (x86)\chromedriver.exe'
```

```python
driver = webdriver.Chrome(driver_path)
driver.get('https://yocket.in/account/login')
time.sleep(2)

email = driver.find_element_by_xpath('//input[@name ="email"]')
email.send_keys('') ### Put in your User Name
submit = driver.find_element_by_xpath('//button[@type="submit"]')
submit.click()
time.sleep(2)

password = driver.find_element_by_xpath('//input[@name ="password"]')
password.send_keys('') ### Put in your Password
submit = driver.find_element_by_xpath('//button[@type="submit"]')
submit.click()
time.sleep(1)

driver.get('https://yocket.in/university-application-deadlines?level=undergrad')
Degree = ['masters','undergrad']
extra = ['Yocket Link','Course Link','University Link']
time.sleep(1)

for z in range(len(Degree)):
    file_name = Degree[z] + "Deadline.csv"
    f = open(file_name,'w')
    page = 1
    Head = driver.find_elements_by_xpath('//div[@id="' + Degree[z] +'"]/table/thead/tr/th')
    Header = [ ii.text for ii in Head]
    Header += extra
    for item in Header:
        if item != Header[-1]:
            f.write(item + ';')
        else:
            f.write(item + '\n')
    time.sleep(5) 
    present = True
    
    while present == True:
#     while page < 79 :
        time.sleep(5)
        rows = driver.find_elements_by_xpath('//div[@id="' + Degree[z] +'"]/table/tbody/tr')
        for j in range(len(rows)):
            cols = driver.find_elements_by_xpath('//div[@id="' + Degree[z] +'"]/table/tbody/tr[' + str(j + 1) + ']/td')
            link = driver.find_elements_by_xpath('//div[@id="' + Degree[z] +'"]/table/tbody/tr[' + str(j + 1) + ']/td//a')
            for i in range(len(cols)):
                if i == 0:
                    f.write(cols[i].text + ';')
                else:
                    f.write(cols[i].text.split('\n')[0] + ';')
            for i in range(len(link)):
                if i != (len(link)-1):
                    f.write (link[i].get_attribute('href')+ ';')
                else:
                    f.write(link[i].get_attribute('href') + '\n')   
                    
        if page/6 ==0:
            time.sleep(120)
        if driver.find_elements_by_xpath('//div[@class="win_close sqico-larrow"]') == True:
            driver.find_elements_by_xpath('//div[@class="win_close sqico-larrow"]').click()
        try:
            driver.find_element_by_xpath('//li[@class="next"]/a').click()
            time.sleep(5)
            try:
                driver.find_element_by_xpath('//li[@class="next"]/a')
                preset = True
            except:
                present = False
                
        except ElementClickInterceptedException:
            time.sleep(5) # Sometimes the pop-up takes time to load
            driver.find_element_by_xpath('//li[@class="next"]/a').click()
        
        
        page = page + 1
        
    f.close()
    
    if z == 0:
        toggle = driver.find_element_by_xpath('//li[@value="'+ Degree[z+1] +'"]')
        toggle.click()
        time.sleep(10)
driver.quit()
```
